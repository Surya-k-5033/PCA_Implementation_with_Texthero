```python
# Import libraries
import pandas as pd
import texthero as hero
```

```python
#reading dataset 
df_fake = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Principal Compound Analysis(PCA)/Fake.csv")
df_fake = df_fake.head(1000)
df_true = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Principal Compound Analysis(PCA)/True.csv")
df_true = df_true.head(1000)
```

```python
# Lable True and False
df_fake['fake'] = 1
df_true['fake'] = 0
```

```python
# concat the file
df = pd.concat([df_fake, df_true], ignore_index = True)
```

```python
# Drop the unwanted column
df = df.drop(['title', 'subject', 'date'], axis = 1)
```

```python
df
```

```python
# Impementing TFIDI and PCA with texthero
df["pca"] = (
            df["text"]
            .pipe(hero.tfidf) 
            .pipe(hero.pca)
)
```

```python
df.head()
```

```python
# Ploting in Scatterplot
hero.scatterplot(
    df, 
    col='pca', 
    color='fake', 
    title="PCA on True or False"
)
```
